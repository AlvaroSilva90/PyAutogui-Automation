```python
import pyautogui as pgui
import pyperclip as pclip
import time as tm
import pandas as pd
from datetime import date
```

```python
def descanso(x):
    tm.sleep(x)
```

```python
pgui.PAUSE = 1

pgui.hotkey("ctrl","t")
pclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pgui.hotkey("ctrl","v")
pgui.press("enter")
    
descanso(2)

pgui.doubleClick(x=-1498, y=131)

descanso(2)

pgui.click(x=-1510, y=208)
pgui.rightClick(x=-1510, y=208)
descanso(2)
pgui.click(x=-1361, y=683)

descanso(5)
```

```python
tbl_Vendas = pd.read_excel(r"A:\Downloaded\Vendas - Dez.xlsx")
# display(tbl_Vendas)
```

```python
# Formatando valores numéricos 
qtdVendFinal = f'{tbl_Vendas["Quantidade"].sum():,}'
qtdVendFinal = qtdVendFinal.replace(',','.')

faturamentoTotal = f'{tbl_Vendas["Valor Final"].sum():_.2f}'
faturamentoTotal= faturamentoTotal.replace('.',',').replace('_','.')

# print(qtdVendFinal,faturamentoTotal)
```

```python
# Abertura de nova aba e inserindo o link para escrever novo email

pgui.hotkey("ctrl","t")
pclip.copy("https://mail.google.com/mail/u/0/#inbox?compose=new")
pgui.hotkey("ctrl","v")
pgui.press("enter")

descanso(8)

# Aponto cursor para inserir o endereço e demais informações do email

pgui.click(x=-504, y=437)

# Adicionando destinatário
pgui.write("ac9271@gmail.com")
pgui.press("tab")
pgui.press("tab")

# Adicionando assunto e data
pclip.copy("Relatório de Vendas ")
pgui.hotkey("ctrl","v")

dataHoje = date.today().strftime('%d/%m/%Y')

pgui.write(str(dataHoje))

pgui.press("tab")

#Adicionando corpo do email
texto = f'''
Prezadas e prezados,

    Segue relatório de vendas de {dataHoje}.
        Quantidade total vendida: {qtdVendFinal}.
        Valor total faturado: $ {faturamentoTotal}.
    
    Quaisquer dúvidas estou à disposição.
    
'''
# Colando o texto e enviando o email
pclip.copy(texto)
pgui.hotkey("ctrl","v")
pgui.hotkey("ctrl","enter")
```
